```python
import torch
import torchvision.transforms as transforms
from datasets import TextDataset
from miscc.config import cfg, cfg_from_file

cfg_from_file("cfg/DAMSM/coco.yml")
imsize = cfg.TREE.BASE_SIZE * (2 ** (cfg.TREE.BRANCH_NUM - 1))

image_transform = transforms.Compose([
        transforms.Resize(int(imsize * 76 / 64)),
        transforms.RandomCrop(imsize),
        transforms.RandomHorizontalFlip()])
dataset = TextDataset(cfg.DATA_DIR, "train",
                      base_size=cfg.TREE.BASE_SIZE,
                      transform=image_transform)
dataloader = torch.utils.data.DataLoader(
    dataset, batch_size=cfg.TRAIN.BATCH_SIZE,
    drop_last=True, shuffle=0, num_workers=int(cfg.WORKERS))
```

```python
# 遍历 dataloader 获取批次数据
data_iter = iter(dataloader)
data = next(data_iter)
print(len(data))
imgs, caps, cap_len, cls_id, key = data
print(f"Images: {imgs[0]}")
print(f"Captions: {caps[0]}")
print(f"Caption Length: {cap_len[0]}")
print(f"Class ID: {cls_id[0]}")
print(f"Key: {key[0]}")

# for i, data in enumerate(dataloader):
#     # 你的 data 通常是一个 tuple, 比如 (images, captions)
#     print(len(data))
#     imgs, caps, cap_len, cls_id, key, caps_two, cap_len_two = data
#     
#     print(f"Batch {i + 1}")
#     print(f"Images: {imgs[0]}")
#     print(f"Captions: {caps[0]}")
#     print(f"Caption Length: {cap_len[0]}")
#     print(f"Class ID: {cls_id[0]}")
#     print(f"Key: {key[0]}")
#     print(f"Captions Two: {caps_two[0]}")
#     print(f"Caption Length Two: {cap_len_two[0]}")
#     
#     if i == 0:
#         break
```
